```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
key_nums_of_As = [1, 13, 25, 37, 49, 61, 73, 85]
freqs_of_As = [27.5, 55, 110, 220, 440, 880, 1760, 3520]

plt.figure(figsize=(7,4))
plt.plot(key_nums_of_As, freqs_of_As, 'o')
plt.xlabel("Key number")
plt.ylabel("Frequency, Hz")
plt.title("Frequencies of the 'A' keys")
plt.savefig('A_plot_1.png')
```

```python
plt.figure(figsize=(7,4))
plt.plot(key_nums_of_As, freqs_of_As, 'o')
plt.xlabel("Key number")
plt.ylabel("Frequency, Hz")
plt.yscale('log')
plt.title("Frequencies of the 'A' keys, with log scale y-axis")
plt.savefig('A_plot_2.png')
```

```python
plt.figure(figsize=(7,4))
plt.plot(key_nums_of_As, freqs_of_As)
plt.plot(key_nums_of_As, freqs_of_As, 'o')
plt.xlabel("Key number")
plt.ylabel("Frequency, Hz")
plt.yscale('log')
plt.title("Frequencies of the 'A' keys, with log scale y-axis")
plt.savefig('A_plot_3.png')

```
